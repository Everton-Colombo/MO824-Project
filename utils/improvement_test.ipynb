```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminhos dos CSVs
csv_first = "agricultural_csp/results_v1/inst_32_large_many_small_with_border_spray20_first_history.csv"
csv_best = "agricultural_csp/results_v1/inst_32_large_many_small_with_border_spray20_best_history.csv"

# L os dois arquivos
df_first = pd.read_csv(csv_first)
df_best = pd.read_csv(csv_best)

# Converte o eixo X de iteraes para tempo (em segundos)
max_iter_first = df_first["Iteration"].max()
max_iter_best = df_best["Iteration"].max()

df_first["Tempo (s)"] = df_first["Iteration"] / max_iter_first * 1800
df_best["Tempo (s)"] = df_best["Iteration"] / max_iter_best * 1800

# Cria figura
plt.figure(figsize=(10, 6))

# Plota apenas a evoluo do melhor objetivo em funo do tempo
plt.plot(df_first["Tempo (s)"], df_first["Best_Obj"], label="First Improving", linewidth=2, color="#1f77b4")
plt.plot(df_best["Tempo (s)"], df_best["Best_Obj"], label="Best Improving", linewidth=2, color="#ff7f0e")

# Personalizao do grfico
plt.title("Comparao da Melhoria da Soluo  First vs Best Improving")
plt.xlabel("Tempo (s)")
plt.ylabel("Melhor Valor do Objetivo")
plt.grid(True, linestyle=":")
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Caminho do CSV
csv_first = "agricultural_csp/results_v1/inst_02_small_none_no_border_spray20_first_history.csv"

# L o CSV
df = pd.read_csv(csv_first)

# Converte iteraes em tempo real (0 a 1800 s)
max_iter = df["Iteration"].max()
df["Tempo (s)"] = df["Iteration"] / max_iter * 1800

# Cria figura base
fig, ax1 = plt.subplots(figsize=(10, 6))
ax2 = ax1.twinx()
ax3 = ax1.twinx()  # terceiro eixo (extra)

# Desloca o terceiro eixo um pouco pra direita (evita sobreposio)
ax3.spines["right"].set_position(("outward", 60))

# Define cores distintas (viridis ou manual)
color_cov = "#1f77b4"   # azul
color_dist = "#2ca02c"  # verde
color_man = "#ff7f0e"   # laranja

# Plota cada componente com seu eixo
p1, = ax1.plot(df["Tempo (s)"], df["Best_Cov_Pen"], color=color_cov, linewidth=2, label="Cobertura Penalizada")
p2, = ax2.plot(df["Tempo (s)"], df["Best_Dist"], color=color_dist, linewidth=2, label="Distncia")
p3, = ax3.plot(df["Tempo (s)"], df["Best_Man_Pen"], color=color_man, linewidth=2, label="Penalizao de Manejo")

# Ajusta limites inferiores para comear em 0
ax1.set_ylim(bottom=0)
ax2.set_ylim(bottom=0)
ax3.set_ylim(bottom=0)

# Configura eixos
ax1.set_xlabel("Tempo (s)")
ax1.set_ylabel("Cobertura Penalizada", color=color_cov)
ax2.set_ylabel("Distncia", color=color_dist)
ax3.set_ylabel("Penalizao de Manejo", color=color_man)

# Ajusta cores dos ticks
ax1.tick_params(axis="y", colors=color_cov)
ax2.tick_params(axis="y", colors=color_dist)
ax3.tick_params(axis="y", colors=color_man)

# Grade leve no eixo principal
ax1.grid(True, linestyle=":", alpha=0.7)

# Legenda combinada (trs curvas)
lines = [p1, p2, p3]
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc="upper right")

# Ttulo e layout
plt.title("Evoluo das Componentes do Objetivo  Estratgia First Improving")
plt.tight_layout()
plt.show()
```
